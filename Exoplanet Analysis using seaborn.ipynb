# source: seaborn library documentation (horizontal boxplot)

```python
%matplotlib inline
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

planets = pd.read_csv("planets.csv", sep = ',')

sns.set_theme(style="ticks")

# Initialize the figure with a logarithmic x axis
f, ax = plt.subplots(figsize=(7, 6))
# if x-scale is set to linear, it becomes difficult to visualize
ax.set_xscale("log")

# Plot the orbital period with horizontal boxes
sns.boxplot(x="st_mass", y="pl_discmethod", data=planets,
            width=.6, palette="vlag")

# Add in points to show each observation
sns.stripplot(x="st_mass", y="pl_discmethod", data=planets,
              size=2, color=".3", linewidth=0)

# Create the visual presentation
ax.xaxis.grid(True)
ax.set(ylabel="")
sns.despine(trim=True, left=True)
```
